```python
import pandas as pd
path = r'D:\py\DH\订单核算\订单列表.xls'
df = pd.read_excel(path)
```

```python
df_valid = df[df['订单状态'] != '订单取消']
df_valid = df_valid[['订单号','订单状态', '订单总额 (US $)', '佣金金额 (US $)', '支付手续费 (US $)', '实收金额 (US $)', '产品信息']]
df_valid['产品编号'] = df_valid['产品信息'].str.extract(r'(\w+-\d+)-\w+')
df_valid['包装'] = df_valid['产品信息'].str.extract(r'\w+-\d+-(\w+)')
df_valid
```

```python
price_dict = {
    "KL42097" : 120,
    "SXZ9175" : 35.8,
    "KL42102" : 118.99,
}
# 产品单价根据产品编号的前面部分匹配
df_valid['单价'] = df_valid['产品编号'].str.extract(r'(\w+)')
df_valid['单价'] = df_valid['单价'].map(price_dict)

pack_price_dict = {
    'Box' : 20,
    'Bag' : 0,
}

df_valid['包装费'] = df_valid['包装'].map(pack_price_dict)
df_valid['总价'] = df_valid['单价'] + df_valid['包装费']
df_valid    
```

```python
# 将订单号列转化为str格式
df_valid['订单号'] = df_valid['订单号'].astype('str')
```

```python
df_shipping = pd.read_excel(r'D:\py\DH\订单核算\运费列表.xlsx')
df_shipping
```

```python
# 将订单号列转化为str格式
df_shipping['订单号'] = df_shipping['订单号'].astype('str')
```

```python
# 根据订单号获取df_shipping中的对应实际运费，如果没有则为空
df_valid['实际运费'] = df_valid['订单号'].map(df_shipping.set_index('订单号')['实际运费'])
df_valid
```

```python
df_valid_has_shipping = df_valid.dropna(subset=['实际运费'])
df_valid_has_shipping['利润'] = df_valid_has_shipping['实收金额 (US $)']*7.2 - df_valid_has_shipping['总价'] - df_valid_has_shipping['实际运费']
df_valid_has_shipping
```

```python
# 计算利润率并转化为2位小数的百分比
df_valid_has_shipping['利润率'] = df_valid_has_shipping['利润'] / (df_valid_has_shipping['实收金额 (US $)']*7.2)
df_valid_has_shipping['利润率'] = df_valid_has_shipping['利润率'].apply(lambda x: format(x, '.2%'))
# 保存为excel
# 如果利润为负数，且产品编号包含KL42102，则修改单价为35.8，重新计算利润
df_valid_has_shipping.loc[(df_valid_has_shipping['利润'] < 0) & (df_valid_has_shipping['产品编号'].str.contains('KL42102')), '单价'] = 35.8
df_valid_has_shipping['总价'] = df_valid_has_shipping['单价'] + df_valid_has_shipping['包装费']
df_valid_has_shipping['利润'] = df_valid_has_shipping['实收金额 (US $)']*7.2 - df_valid_has_shipping['总价'] - df_valid_has_shipping['实际运费']
df_valid_has_shipping['利润率'] = df_valid_has_shipping['利润'] / (df_valid_has_shipping['实收金额 (US $)']*7.2)
df_valid_has_shipping['利润率'] = df_valid_has_shipping['利润率'].apply(lambda x: format(x, '.2%'))
```

```python
df_valid_has_shipping.to_excel(r'D:\py\DH\订单核算\订单利润0729.xlsx', index=False)
```
